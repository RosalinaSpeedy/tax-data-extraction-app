```python
import re

class W2FieldPatterns:
    """
    Holds regex patterns for W-2 form fields and provides simple extraction utilities.
    """

    def __init__(self):
        # Define your label anchors (for matching in text)
        self.patterns = [
            "Employee's social security number",
            "Employer identification number",
            "OMB No.",
            "Control number",
            "Employer's name, address, and ZIP code",
            "Employee's first name and initial Last name",
            "Employee's address and ZIP code",
            "Wages, tips, other compensation",
            "Federal income tax withheld",
            "Social security wages",
            "Social security tax withheld",
            "Medicare wages and tips",
            "Medicare tax withheld",
            "Social security tips",
            "Allocated tips",
            "Advance EIC payment",
            "Dependent care benefits",
            "Nonqualified plans",
            "See instructions for box 12",
            "Statutory employee",
            "Retirement plan",
            "Third-party sick pay",
            "Other (see enclosed Notice to Employee)",
            "Employer's state ID number",
            "State wages, tips, etc",
            "State income tax",
            "Local wages, tips, etc",
            "Local income tax",
            "Locality name",
            "Department of the Treasury—Internal Revenue Service",
            "Form W-2",
            "Copy B—To Be Filed with Employee's FEDERAL Tax Return",
            "Copy C—For EMPLOYEE'S RECORDS",
            "Wage and Tax Statement"
        ]

        # Define regex extraction patterns for numeric/text fields
        self.field_patterns = {
    # Identification fields
    "Employee's social security number": r"(?:\b|_)(\d{3}[-\s]?\d{2}[-\s]?\d{4})(?:\b|_)",
    "Employer identification number": r"(?:\b|_)(\d{2}[-\s]?\d{7})(?:\b|_)",
    "Control number": r"Control(?:\s+number)?[:\s]*([A-Z0-9-]{4,})",

    # Employer / Employee text fields
    "Employer's name, address, and ZIP code": r"(?:(?:Employer|Employers)[’'`s]*\s*name[^\n]+)([A-Za-z0-9,.\s-]+[A-Z]{2}\s*\d{5}(?:-\d{4})?)",
    "Employee's first name and initial Last name": r"(?:Employee[’'`s]*\s*(?:first\s*name.*?))([A-Z][a-z]+\s+[A-Z][a-z]+)",
    "Employee's address and ZIP code": r"([0-9]+\s+[A-Za-z\s]+[A-Z]{2}\s*\d{5}(?:-\d{4})?)",

    # Monetary fields (tolerate commas, OCR missing dots, etc.)
    "Wages, tips, other compensation": r"Wages[^\d]*(\d[\d,]*\.?\d{0,2})",
    "Federal income tax withheld": r"Federal[^\d]*(\d[\d,]*\.?\d{0,2})",
    "Social security wages": r"Social\s+security\s+wages[^\d]*(\d[\d,]*\.?\d{0,2})",
    "Social security tax withheld": r"Social\s+security\s+tax[^\d]*(\d[\d,]*\.?\d{0,2})",
    "Medicare wages and tips": r"Medicare\s+wages[^\d]*(\d[\d,]*\.?\d{0,2})",
    "Medicare tax withheld": r"Medicare\s+tax[^\d]*(\d[\d,]*\.?\d{0,2})",
    "Social security tips": r"Social\s+security\s+tips[^\d]*(\d[\d,]*\.?\d{0,2})",
    "Allocated tips": r"Allocated\s+tips[^\d]*(\d[\d,]*\.?\d{0,2})",
    "Advance EIC payment": r"Advance\s+EIC[^\d]*(\d[\d,]*\.?\d{0,2})",
    "Dependent care benefits": r"Dependent\s+care[^\d]*(\d[\d,]*\.?\d{0,2})",
    "Nonqualified plans": r"Nonqualified\s+plans[^\d]*(\d[\d,]*\.?\d{0,2})",

    # State / Local fields
    "Employer's state ID number": r"[A-Z]{2}\s*\d{3}[-\s]?\d{2}[-\s]?\d{2,4}",
    "State wages, tips, etc": r"State\s+wages[^\d]*(\d[\d,]*\.?\d{0,2})",
    "State income tax": r"State\s+income\s+tax[^\d]*(\d[\d,]*\.?\d{0,2})",
    "Local wages, tips, etc": r"Local\s+wages[^\d]*(\d[\d,]*\.?\d{0,2})",
    "Local income tax": r"Local\s+income\s+tax[^\d]*(\d[\d,]*\.?\d{0,2})",
    "Locality name": r"Locality\s+name\s*([A-Za-z\s]+)",
}


    def extract_field(self, text: str, field_label: str):
        """
        Extracts the value of a given field from text using regex.
        """
        pattern = self.field_patterns.get(field_label)
        if not pattern:
            return None
        match = re.search(pattern, text)
        return match.group() if match else None

    def extract_all(self, text: str):
        """
        Runs extraction for all known fields and returns a dict.
        """
        results = {}
        for label in self.field_patterns:
            val = self.extract_field(text, label)
            if val:
                results[label] = val
        return results

    def list_fields(self):
        """Returns all supported field labels."""
        return list(self.field_patterns.keys())
```

```python
extractor = W2FieldPatterns()
results = extractor.extract_all(text)

from pprint import pprint
pprint(results)
```
