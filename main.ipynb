```python
import pymupdf as fitz
from pdf2image import convert_from_path
import pytesseract
```

```python
doc_path = '/W2_XL_input_clean_1000.jpg'
text = ""
if doc_path.endswith('.pdf'):
    doc = convert_from_path(doc_path)
    for page in doc:
        text += pytesseract.image_to_string(page)
else:
    doc = doc_path
    text += pytesseract.image_to_string(doc)
#doc = convert_from_path('/W2_XL_input_clean_1000.pdf')
print(text)
```
